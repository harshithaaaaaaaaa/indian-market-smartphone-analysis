```python
import pandas as pd

# Load dataset
file_path = r'C:\Users\Harshitha\projects\indian_market_smartphone_analysis\smartphone_cleaned_v5.csv'
df = pd.read_csv(file_path)

# Show basic info and first 5 rows
print("Dataset shape:", df.shape)
print("\nColumns:\n", df.columns)
print("\nSample data:\n", df.head())
```

```python
# Check missing values
print("Missing values per column:")
print(df.isnull().sum())

# Convert price and rating columns to numeric if needed (replace 'Price' and 'Rating' with actual column names)
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Drop rows where Price or Rating is missing
df = df.dropna(subset=['Price', 'Rating'])

# Create price segments (you can adjust these thresholds)
def price_segment(price):
    if price < 10000:
        return 'Budget'
    elif price < 25000:
        return 'Mid-Range'
    else:
        return 'Premium'

df['Price Segment'] = df['Price'].apply(price_segment)

# Summary statistics
print("\nData Summary:")
print(df[['Price', 'Rating']].describe())

print("\nPrice Segment counts:")
print(df['Price Segment'].value_counts())
```

```python
# Check missing values
print("Missing values per column:")
print(df.isnull().sum())

# Convert price and rating columns to numeric (if not already)
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Drop rows where price or rating is missing
df = df.dropna(subset=['price', 'rating'])

# Create price segments
def price_segment(price):
    if price < 10000:
        return 'Budget'
    elif price < 25000:
        return 'Mid-Range'
    else:
        return 'Premium'

df['price_segment'] = df['price'].apply(price_segment)

# Summary statistics
print("\nData Summary:")
print(df[['price', 'rating']].describe())

print("\nPrice Segment counts:")
print(df['price_segment'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of Price segments
plt.figure(figsize=(7,5))
sns.countplot(data=df, x='price_segment', palette='viridis')
plt.title('Count of Smartphones by Price Segment')
plt.xlabel('Price Segment')
plt.ylabel('Number of Models')
plt.show()

# Average rating by Price Segment
plt.figure(figsize=(7,5))
sns.barplot(data=df, x='price_segment', y='rating', palette='viridis')
plt.title('Average Rating by Price Segment')
plt.xlabel('Price Segment')
plt.ylabel('Average Rating')
plt.show()

# Brand-wise average rating (top 10 brands)
top_brands = df['brand_name'].value_counts().nlargest(10).index
plt.figure(figsize=(10,6))
sns.barplot(data=df[df['brand_name'].isin(top_brands)], 
            x='brand_name', y='rating', palette='magma')
plt.title('Average Rating of Top 10 Brands')
plt.xlabel('Brand')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
# 1. Correlation heatmap (see which features impact price and rating)
plt.figure(figsize=(12,8))
sns.heatmap(df[['price', 'rating', 'ram_capacity', 'battery_capacity', 'screen_size', 'refresh_rate']].corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()

# 2. Price vs Rating scatterplot with trendline
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='price', y='rating', hue='price_segment', palette='viridis')
sns.regplot(data=df, x='price', y='rating', scatter=False, color='red')
plt.title('Price vs Rating with Trendline')
plt.show()

# 3. Average battery capacity by price segment
plt.figure(figsize=(7,5))
sns.barplot(data=df, x='price_segment', y='battery_capacity', palette='Blues')
plt.title('Average Battery Capacity by Price Segment')
plt.show()

# 4. Number of smartphones with 5G by price segment
plt.figure(figsize=(7,5))
sns.countplot(data=df, x='price_segment', hue='has_5g', palette='Set2')
plt.title('5G Availability across Price Segments')
plt.show()
```
